Data Preprocessing

```python
df['State'].values
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder=OneHotEncoder(sparse=False)
encoded_state=encoder.fit_transform(df['State'].values.reshape(-1,1))
encoded_state
print(encoded_state)

#Create new data frame with encoded_states
encoded_df=pd.DataFrame(encoded_state,columns=encoder.categories_[0])
encoded_df

#Concatanate original df with encoded_df
df=pd.concat([df,encoded_df],axis=1)
print(df)
```
